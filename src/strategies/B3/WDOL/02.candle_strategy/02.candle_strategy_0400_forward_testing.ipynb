```python
def is_short(current_close, bars, target, stop):
    correct_target = current_close - target
    correct_stop = current_close + stop
    
    for x in bars:
        if x['high'] >= correct_stop:
            return 0
        if x['low'] < correct_target:
            return 1
    return 0
    
def is_long(current_close, bars, target, stop):
    correct_target = current_close + target
    correct_stop = current_close - stop
    
    for x in bars:
        if x['high'] >= correct_target:
            return 1
        if x['low'] < correct_stop:
            return 0
    return 0
```

```python
def get_slope(current_bar, value_to_calculate):
    a = []
    b = []
    for current_x in range(0,11):
        a.append(current_x)
        b.append(current_bar[f'x{current_x}_{value_to_calculate}'])
    return linregress(a,b).slope
```

```python
def is_short_signal(current_date_time, results, current_candle, indicators):
    if results[current_date_time]['short_predict'] < results[current_date_time]['long_predict']:
        return False
    return results[current_date_time]['short_predict'] >= DECISION_BOUNDARY
        
def is_long_signal(current_date_time, results, current_candle, indicators):
    if results[current_date_time]['short_predict'] > results[current_date_time]['long_predict']:
        return False
    return results[current_date_time]['long_predict'] >= DECISION_BOUNDARY
    
```

```python
current_max = 0
skip_candles = 0

def calculate_max_gain_short(current_trade, current_max):
    return (current_trade['trade_start']- current_max) / (current_trade['trade_start'] - current_trade['trade_target'])
    
def calculate_max_gain_long(current_trade, current_max):
    return (current_max - current_trade['trade_start']) / (current_trade['trade_target'] - current_trade['trade_start'])    

for current_candle in current_raw_dataset:
    if skip_candles > 0:
        skip_candles = skip_candles - 1
        continue

    current_date_time = (current_candle['Date'] * 10000) + current_candle['Time']
    indicators = [x for x in generated_bars if x['current_date'] == current_candle['Date'] and x['current_time']==current_candle['Time']]
    if len(indicators) == 0:
        continue
    
    if current_trade is not None:
        is_short = current_trade['trade_type'] == 'short'
        is_long = current_trade['trade_type'] == 'long'
        if USE_SMART_STOP:
            if (is_long and is_short_signal(current_date_time, results, current_candle, indicators)) or \
               (is_short and is_long_signal(current_date_time, results, current_candle, indicators)):
                current_trade['smart_stop'] = True
        if is_short:
            if current_candle['low'] < current_max:
                current_max = current_candle['low']
                if current_max < current_trade['trade_target']:
                    current_max = current_trade['trade_target']       
                    
            if calculate_max_gain_short(current_trade, current_max) > 0.88:
                current_trade['trade_stop']=current_trade['trade_start'] - ((current_trade['trade_start']-current_trade['trade_target']) / 1)
                pass
                    
            if current_candle['low'] <= current_trade['trade_target'] or current_trade['smart_stop']:
                if current_trade['smart_stop']:
                    current_trade['result'] = round(current_trade['trade_start'] - current_candle['close'],2)
                    current_trade['final_close'] = current_candle['close']
                else:
                    current_trade['result'] = round(current_trade['trade_start'] - current_trade['trade_target'],2)
                    current_trade['final_close'] = current_trade['trade_target']
                current_trade['end_time'] = current_candle['Time']
                current_trade['end_candle'] = candle_count
                current_trade['num_candles'] = current_trade['end_candle'] - current_trade['start_candle']
                current_trade['max_gain'] = calculate_max_gain_short(current_trade, current_max)
                current_trade = None
                #skip_candles = 10
                continue
            if current_candle['high'] >= current_trade['trade_stop'] or current_trade['smart_stop']:
                if current_trade['smart_stop']:
                    current_trade['result'] = round(current_trade['trade_start'] - current_candle['close'],2)
                    current_trade['final_close'] = current_candle['close']
                else:
                    current_trade['result'] = round(current_trade['trade_start'] - current_trade['trade_stop'],2)
                    current_trade['final_close'] = current_trade['trade_stop']
                current_trade['end_time'] = current_candle['Time']
                current_trade['end_candle'] = candle_count
                current_trade['num_candles'] = current_trade['end_candle'] - current_trade['start_candle']
                current_trade['max_gain'] = calculate_max_gain_short(current_trade, current_max)
                current_trade = None
                #skip_candles = 10
                continue                
        if is_long:
            if current_candle['high'] > current_max:
                current_max = current_candle['high']
                if current_max > current_trade['trade_target']:
                    current_max = current_trade['trade_target']

            if calculate_max_gain_short(current_trade, current_max) > 0.88:
                current_trade['trade_stop']=current_trade['trade_start'] + ((current_trade['trade_target']-current_trade['trade_start']) / 1)
                pass
                    
            if current_candle['low'] <= current_trade['trade_stop'] or current_trade['smart_stop']:
                if current_trade['smart_stop']:
                    current_trade['result'] = round(current_candle['close'] - current_trade['trade_start'],2)
                    current_trade['final_close'] = current_candle['close']
                else:
                    current_trade['result'] = round(current_trade['trade_stop'] - current_trade['trade_start'],2)
                    current_trade['final_close'] = current_trade['trade_stop']
                current_trade['end_time'] = current_candle['Time']
                current_trade['end_candle'] = candle_count
                current_trade['num_candles'] = current_trade['end_candle'] - current_trade['start_candle']
                current_trade['max_gain'] = calculate_max_gain_long(current_trade, current_max)
                current_trade = None
                #skip_candles = 10
                continue
            if current_candle['high'] >= current_trade['trade_target'] or current_trade['smart_stop']:
                if current_trade['smart_stop']:
                    current_trade['result'] = round(current_candle['close'] - current_trade['trade_start'],2)
                    current_trade['final_close'] = current_candle['close']                    
                else:
                    current_trade['result'] = round(current_trade['trade_target'] - current_trade['trade_start'],2)
                    current_trade['final_close'] = current_trade['trade_target']
                current_trade['end_time'] = current_candle['Time']
                current_trade['end_candle'] = candle_count
                current_trade['num_candles'] = current_trade['end_candle'] - current_trade['start_candle']
                current_trade['max_gain'] = calculate_max_gain_long(current_trade, current_max)
                current_trade = None
                #skip_candles = 10
                continue        
                
        if candle_count > (current_trade['start_candle'] + MAX_TRADE_DURATION):
            if is_long:
                current_trade['result'] = round(current_candle['close'] - current_trade['trade_start'],2)
                current_trade['final_close'] = current_candle['close']
                current_trade['end_time'] = current_candle['Time']
                current_trade['end_candle'] = candle_count
                current_trade['num_candles'] = current_trade['end_candle'] - current_trade['start_candle']
                current_trade['max_gain'] = calculate_max_gain_long(current_trade, current_max)
            if is_short:
                current_trade['result'] =  round(current_trade['trade_start'] - current_candle['close'],2)
                current_trade['final_close'] = current_candle['close']
                current_trade['end_time'] = current_candle['Time']
                current_trade['end_candle'] = candle_count
                current_trade['num_candles'] = current_trade['end_candle'] - current_trade['start_candle']
                current_trade['max_gain'] = calculate_max_gain_short(current_trade, current_max)
            current_trade = None
            continue
        
    if current_trade is None:
        if current_candle['Date'] > TEST_MINIMUM_DATE_TRADE and \
           current_candle['Time'] >= MINIMUM_TIME and \
           current_candle['Time'] <= MAXIMUM_TIME:
            is_short = is_short_signal(current_date_time, results, current_candle, indicators)
            is_long = is_long_signal(current_date_time, results, current_candle, indicators)
            if is_short or is_long:
                current_trade = {**current_candle, 
                                 'trade_type': 'short' if is_short else 'long', 
                                 'start_candle': candle_count,
                                 'trade_start': current_candle['close'],
                                 'predicted': results[current_date_time]['short_predict'] if is_short else results[current_date_time]['long_predict'],
                                 'shap': results[current_date_time]['short_shap'] if is_short else results[current_date_time]['long_shap'],
                                 'smart_stop': False
                                }
                if is_short:
                    current_trade['trade_target'] = current_candle['close'] * (1 - (current_target / 100))
                    current_trade['trade_stop'] = current_candle['close'] * (1 + (current_stop / 100))
                    current_max = current_candle['close']
                if is_long:
                    current_trade['trade_target'] = current_candle['close'] * (1 + (current_target / 100))
                    current_trade['trade_stop'] = current_candle['close'] * (1 - (current_stop / 100))
                    current_max = current_candle['close']
                
                processed_dates[current_candle['Date']] = '1'
                current_trade_entries.append(current_trade)
                
    candle_count += 1
```
