# Create Dataframes

This notebook is responsible for generating the dataframes that will be used for the training of the models.

The API uses data from the last 11 candles and several indicator like:
* Exponential Moving Averages (9,21,55,144,233)
* Volume Weighted Average Price (VWAP)
* Last Trading day: Open, High, Low, Close

The candle's values and their indicators are normalized as the percent change of the close value for the candle that this dataframe belongs to.

So, if the reference candle close is 5500 points, and their indicators are:
* Last Day High: 5600
* Last Day Open: 5700
* Exponential MM 21 for the 1st Candle: 5400
* Exponential MM 55 for the 1st Candle: 5450

It will be recorded in the dataframe as:
* Last Day High: 1.81
* Last Day Open: 3.63
* Exponential MM 21 for the 1st Candle: -1.81
* Exponential MM 55 for the 1st Candle: -0.917

The functions below determine if we have reached our target or if we had to stop our operation.

```python
def is_short(current_close, bars, target, stop):
    correct_target = current_close - target
    correct_stop = current_close + stop
    
    for x in bars:
        if x['high'] >= correct_stop:
            return 0
        if x['low'] < correct_target:
            return 1
    return 0
    
def is_long(current_close, bars, target, stop):
    correct_target = current_close + target
    correct_stop = current_close - stop
    
    for x in bars:
        if x['high'] >= correct_target:
            return 1
        if x['low'] < correct_stop:
            return 0
    return 0
```

It is important to notice that all data will be trained "normalized", so we don't train XGBoost with the absolute values, but with the percent changes in relation to the open of the referenced candle. 

So, for example, if the close of the candle that we are calculating is 4900 points and the previous close was 4950 points, the training data for the previous close will be 1.02, which represents a change of 1.02% in relation to the current candle open.

All data that is used for training is percent-base in order to better generalize the inference algorithm.
